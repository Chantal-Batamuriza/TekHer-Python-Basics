### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.


```python
# prompt: Plot a contour plot for the function z = sin(x) + cos(y) over a grid of x and y values ranging from -5 to 5.

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values
Z = np.sin(X) + np.cos(Y)

# Create the contour plot
plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, cmap='viridis') # You can change the colormap
plt.colorbar(label='z = sin(x) + cos(y)') # Add a color bar
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of z = sin(x) + cos(y)')
plt.show()
```

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.


```python
# prompt: Create a 3D plot of the function z = x^2 + y^2 over a grid of x and y values.

from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Plot of z = x^2 + y^2')

plt.show()
```

### Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.


```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.rand(5)
values = np.append(values, values[0])
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, values, color='blue', linewidth=2)
ax.fill(angles, values, color='blue', alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_ylim(0, 1)

plt.title("Radar Chart with 5 Random Variables")
plt.show()
```

### Exercise 34:
Create a plot with twin y-axes where one axis represents `y = x` and the other represents `y = x^2`.


```python
# prompt: Create a plot with twin y-axes where one axis represents y = x and the other represents y = x^2

# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot y = x on the first y-axis (ax1)
x = np.arange(0, 11)
y1 = x
ax1.plot(x, y1, color='blue', label='y = x')
ax1.set_xlabel('x')
ax1.set_ylabel('y = x', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Create a second y-axis that shares the same x-axis
ax2 = ax1.twinx()

# Plot y = x^2 on the second y-axis (ax2)
y2 = x**2
ax2.plot(x, y2, color='red', label='y = x^2')
ax2.set_ylabel('y = x^2', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add a title to the plot
plt.title('Twin Y-Axes Plot: y = x and y = x^2')

# Add legends for both lines
fig.legend(loc="upper left", bbox_to_anchor=(0.15, 0.85))

plt.show()
```

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# prompt: Create a waterfall chart using Matplotlib to show the changes in a sequence of values.

# Data for the waterfall chart
# Start with an initial value, then a sequence of changes
values = [100, 20, -30, 45, -15, 120]
labels = ['Start', 'Increase A', 'Decrease B', 'Increase C', 'Decrease D', 'End']

# Calculate cumulative values
cumulative_values = np.cumsum(values)
# The last value should represent the final total
cumulative_values[-1] = cumulative_values[-2] + values[-1] # Ensure the last bar represents the final total

# Calculate the bottom of each bar
bottoms = np.zeros(len(values))
# For intermediate bars, the bottom is the previous cumulative value
bottoms[1:] = cumulative_values[:-1]

# Create the figure and axes
plt.figure(figsize=(10, 6))

# Plot the bars
# Use an offset for positive values to start from the previous total
plt.bar(labels, values, bottom=bottoms, color=['skyblue' if v > 0 else 'lightcoral' for v in values])

# Add lines connecting the bars to show the flow (optional but common in waterfall charts)
for i in range(len(values) - 1):
    plt.plot([i, i + 1], [cumulative_values[i], cumulative_values[i]], color='gray', linestyle='--')


# Add annotations (values) above/below the bars
for i, val in enumerate(values):
    # Adjust text position based on whether the value is positive or negative
    text_position = cumulative_values[i] if val > 0 else cumulative_values[i] + val
    plt.text(i, text_position, f'{val:+}', ha='center', va='bottom' if val > 0 else 'top')


# Add the final total annotation
plt.text(len(values) - 1, cumulative_values[-1], f'{cumulative_values[-1]:.2f}', ha='center', va='bottom', fontweight='bold')


# Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Waterfall Chart')
plt.grid(axis='y', linestyle='--')

plt.show()
```

### Exercise 49:
Create a radar chart to visualize the strengths of five different attributes.


```python
# prompt: Create a radar chart to visualize the strengths of five different attributes

# Data for the radar chart
labels = np.array(['Attribute 1', 'Attribute 2', 'Attribute 3', 'Attribute 4', 'Attribute 5'])
# Replace with your actual attribute strength data (values between 0 and 1 or scaled appropriately)
# Example with sample data:
stats = np.array([0.8, 0.6, 0.9, 0.7, 0.5])

# Number of variables
num_vars = len(labels)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Complete the circle for plotting
stats = np.concatenate((stats,[stats[0]]))
angles = np.concatenate((angles,[angles[0]]))

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot data
ax.plot(angles, stats, 'o-', linewidth=2)

# Fill area
ax.fill(angles, stats, alpha=0.25)

# Set the grid labels (variable names)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)

# Set the title
ax.set_title("Attribute Strength Radar Chart")

# Set the extent of the radial axis (adjust based on your data range)
ax.set_ylim(0, 1) # Assuming data is scaled between 0 and 1

ax.grid(True)

plt.show()
```
